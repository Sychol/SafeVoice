```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import matplotlib.font_manager as fm
```

```python
# ✅ 한글 폰트 설정 (윈도우 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'   # 윈도우
# plt.rcParams['font.family'] = 'Pretendard-Bold'   # 윈도우
# plt.rcParams['font.family'] = 'AppleGothic'   # 맥 사용 시 주석 해제!
plt.rcParams['axes.unicode_minus'] = False      # 마이너스 깨짐 방지
```

```python
# ✅ 데이터
data = {
    "precision": [0.85, 0.72, 0.74, 0.87],
    "recall": [0.87, 0.63, 0.77, 0.83],
    "f1-score": [0.86, 0.67, 0.76, 0.85],
    "support": [967, 644, 745, 224]
}
labels = ["혐오", "악플/욕설", "clean", "학교폭력"]
df = pd.DataFrame(data, index=labels)

# ✅ heatmap 그리기
plt.figure(figsize=(8, 6))
ax = sns.heatmap(
    df.iloc[:, :-1],       # support 제외
    annot=True,
    fmt=".2f",
    cmap="YlGnBu",
    cbar=True,
    annot_kws={"size": 15, "weight": "bold"},
)

# ✅ 라벨 폰트 스타일 적용
ax.set_yticklabels(ax.get_yticklabels(), fontsize=18, weight='bold')
ax.set_xticklabels(ax.get_xticklabels(), fontsize=18, weight='bold')
plt.title("학교폭력 분석 BERT 모델", fontsize=25, weight='bold')

# 컬러바 객체 꺼내기
colorbar = ax.collections[0].colorbar

# 컬러바 폰트 변경
colorbar.ax.tick_params(labelsize=15)  # 글자 크기

plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
